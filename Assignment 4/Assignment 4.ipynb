```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
pi=np.pi
x= np.linspace(-pi, pi, 256) 
y1 = np.tan(x)
y2 = np.cos(x)
y3 = np.sin(x)
plt.plot(x,y1,color="r",label="tan")
plt.plot(x,1/y1,color="b",label="cot")
plt.plot(x,y2,color="g",label="sec")
plt.plot(x,y3,color="y",label="cosec")
plt.axis([-5, 5, -2, 2])
plt.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
result1=[2,5,8,5]
result2=[3,2,5,7]
barWidth =0.25
br1=np.arange(len(result1))
br2=[x+barWidth  for x in br1]
ax.bar(br1 , result1, color = 'b', width = barWidth )
ax.bar(br2 , result2, color = 'g', width = barWidth )
plt.xticks([r + barWidth/2 for r in range(len(result1))],
        ['A', 'B', 'C', 'D'])
#ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)
plt.show()
```
